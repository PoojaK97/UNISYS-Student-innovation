```python
import numpy as np
import os
import sys
```

```python
BASE_DIR = ''
TEXT_DATA_DIR = os.path.join(BASE_DIR, './data/')
SUM_DATA_DIR = os.path.join(BASE_DIR, './summary/')
```

```python
texts = []
```

```python
for name in sorted(os.listdir(TEXT_DATA_DIR)):
    fpath = os.path.join(TEXT_DATA_DIR, name)
    args = {} if sys.version_info < (3,) else {'encoding': 'latin-1'}
    with open(fpath, **args) as f:
        t = f.read()
        texts.append(t)
```

```python
max_len = 0
min_len = 1000000
```

```python
listi = []
```

```python
for i in texts:
    size = len(i.split(" "))
    if(size<min_len):
        print("New min is",end=" ")
        print(size)
        min_len = size
    if(max_len<size):
        print("New max is",end=" ")
        print(size)
        max_len = size
    listi.append(size)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.distplot(listi)
```

```python
count1 = 0
count2 = 0
lis2 = []
for j in listi:
    if(j>7700):
        count1 += 1
    elif(j<500):
        count2 += 1
    else:
        lis2.append(j)

sns.distplot(lis2)
print(count1)
print(count2)
```

```python
for name in sorted(os.listdir(TEXT_DATA_DIR)):
    fpath = os.path.join(TEXT_DATA_DIR, name)
    spath = os.path.join(SUM_DATA_DIR, name)
    args = {} if sys.version_info < (3,) else {'encoding': 'latin-1'}
    with open(fpath, **args) as f:
        t = f.read()
        texts.append(t)
        size = len(texts[-1].split(" "))
        if(size>7700 or size<500):
            os.remove(spath)
            os.remove(fpath)
            print("Successfully removed")
        f.close()
```
